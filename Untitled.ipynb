# Simulating simple queues

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Printing arrival time distribution

```python
inter_arrival_time = np.array([1 for _ in range(0,10)])
%store inter_arrival_time

plt.title("Arrival time distribution")
plt.xlabel("Value")
plt.ylabel("Count")
plt.hist(inter_arrival_time, alpha=0.5)
plt.show()
print(f'Mean arrival time: {inter_arrival_time.mean()}')
```

```python
from src.queue import Queue

queue = Queue([1 for _ in range(0,10)], [2 for _ in range(0,10)])
queue.process()

plt.title("Queue size")
plt.xlabel("Time")
plt.ylabel("Size")
plt.plot(queue.wait_times.tolist(), alpha=0.5)
plt.show()
print(f'Mean queue size: {queue.wait_times.mean()}')

plt.show()
```

Departure time and distribution

```python

plt.title("Departure times")
plt.xlabel("Time")
plt.ylabel("Duration")
plt.plot(queue.departure_times, alpha=0.5)
plt.show()

plt.title("Departure distribution")
plt.xlabel("Value")
plt.ylabel("Count")
plt.hist(queue.departure_times, alpha=0.5)
```
